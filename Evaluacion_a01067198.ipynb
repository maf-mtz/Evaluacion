## **Eliminar outliers**

### California

```python
df_Cal = calif.select_dtypes(include=np.number)
df_Cal
```

```python
df_Cal.isnull().sum()
```

```python
df_Cal = df_Cal.drop(['neighbourhood_group_cleansed','bathrooms', 'calendar_updated'], axis=1)
df_Cal
```

```python
df_Cal.isnull().sum()
```

```python
fig = plt.figure(figsize =(15, 8))
df_Cal.plot(kind='box', vert=False)
plt.title("Valores Atípicos de California")
plt.show() #dibujamos el diagrama
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df_Cal

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= df_Cal[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
Cal_iqr= df_Cal[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
Cal_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Cal_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
Cal_clean=Cal_iqr.copy()
Cal_clean=Cal_clean.fillna(round(Cal_iqr.mean(),1))
Cal_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Cal_clean.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Cal_clean.to_csv("California_limpios.csv")
#descargar archivo filtrado en csv
from google.colab import files    
files.download("California_limpios.csv")
```
